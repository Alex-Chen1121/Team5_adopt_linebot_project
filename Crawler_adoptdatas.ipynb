```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 抓到第幾頁從這邊設定
df= pd.DataFrame()
for i in range(3):
# 領養網址
  url=f'https://wepet.tw/%E5%AF%B5%E7%89%A9%E9%A0%98%E9%A4%8A?page={i+1}'
# 憑證
  headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
         ,"Referer":"https://wepet.tw/","Content-Type":"application/json; charset=UTF-8"}
# 網頁資訊
  html=requests.get(url, headers=headers)
# 取得的回應
  response=BeautifulSoup(html.text)
# 創建空的dataframe

  pet_datas= response.find_all("a",class_="text-white text-capitalize d-block text-truncate")
  for name_data in pet_datas:
      sub_url= name_data["href"]
      data=requests.get(sub_url, headers=headers)
      response2=BeautifulSoup(data.text)
      pet_detail=response2.find("div", class_="row font-size-15 mb-3")
      # print(pet_datas)
      species= pet_detail.find_all("div", class_="col-12 mb-1")
      img=response2.find("img", class_="img-fluid")["src"]
      detail=[]
      addition_data=[["名稱",name_data.text],["圖片",img],["資訊網址",sub_url]]
      detail.extend(addition_data)
      for i in species:
          detail.append(i.text.split("："))
          x=dict(detail[0:11])
      df= df.append([x], ignore_index=True)
```

```python
json_str= df.to_json(orient="columns",force_ascii=False)
json_str
```

```python
# 儲存成json檔
import json
filename="領養寵物資訊.json"
with open(filename, 'w') as f_obj:
  json.dump(json_str, f_obj)
```
